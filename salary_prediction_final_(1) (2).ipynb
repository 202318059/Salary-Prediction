```python
def extract_min_salary(salary_string):
    try:
        return int(salary_string.split("-")[0])
    except (ValueError, IndexError):
        return None

df["min_salary"] = salary.apply(extract_min_salary)
df["min_salary"].tail()
```

```python
def title_simplifier(title):
  if 'data scientist' in title.lower():
    return 'data scientist'
  elif 'data engineer' in title.lower():
    return 'data engineer'
  elif 'analyst' in title.lower():
    return 'analyst'
  elif 'machine learning' in title.lower():
    return 'mle'
  elif 'manager' in title.lower():
    return 'manager'
  elif 'director' in title.lower():
    return 'director'
  else:
    return 'na'

df['job_title_simplified'] = df['Job Title'].apply(title_simplifier)
```

```python
df['job_title_simplified'].value_counts()
```

```python
def seniority (title):
  if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower():
    return 'senior'
  elif 'jr' in title.lower() or 'jr.' in title.lower():
    return 'jr'
  else:
    return 'na'

df['job seniority']=df['Job Title'].apply(seniority)
```

```python
df['job seniority'].value_counts()
```

```python
df['Competitors'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x!='-1' else 0)
df['Competitors']
```

```python
df['Type of ownership'].value_counts()
```

```python
def ownership_simplifier(text):
  if 'private' in text.lower():
    return 'Private'
  elif 'public' in text.lower():
    return 'Public'
  elif ('-1' in text.lower()) or ('unknown' in text.lower()):
    return 'Other Organization'
  else:
    return text

df['Type of ownership'] = df['Type of ownership'].apply(ownership_simplifier)
```

```python
df['Type of ownership'].value_counts()
```

```python
df['Revenue'].value_counts()
```

```python
def revenue_simplifier(text):
  if '-1' in text.lower():
    return 'Unknown / Non-Applicable'
  else:
    return text

df['Revenue'] = df['Revenue'].apply(revenue_simplifier)
```

```python
df['Revenue'].value_counts()
```

```python
df['Size'].value_counts()
```

```python
def size_simplifier(text):
  if '-1' in text.lower():
    return 'Unknown'
  else:
    return text

df['Size'] = df['Size'].apply(size_simplifier)
```

# Feature Engineering

Trimming Columns

```python
industry_list = [
    'Aerospace & Defense',
    'Health Care Services & Hospitals',
    'Advertising & Marketing',
    'Consulting',
    'Internet',
    'Others',
    'Biotech & Pharmaceuticals',
    'Enterprise Software & Network Solutions',
    'Insurance Carriers',
    'IT Services',
    'Computer Hardware & Software',
    'Consumer Products Manufacturing'
]
def industry_simplifier(text):
  if text not in industry_list:
    return 'Others'
  else:
    return text
df['Industry']=df['Industry'].apply(industry_simplifier)
```

```python
job_state_list=['CA','MA','NY','VA','IL','MD','PA',]
def job_state_simplifier(text):
  if text not in job_state_list:
    return 'Others'
  else:
    return text
df['job_state']=df['job_state'].apply(job_state_simplifier)
```

```python
df['job_in_headquarters']=df.apply(lambda x: 1 if x['Location']==x['Headquarters'] else 0, axis=1)
```

```python
df_model=df.copy(deep=True)
df_model =df_model[[ 'average-salary','Rating',
        'Type of ownership', 'Industry', 'Revenue', 'Competitors',
        'Size', 'job_state', 'company_age', 'python_job',
       'spark_job', 'aws_job', 'excel_job', 'job_title_simplified',
       'job seniority', 'job_in_headquarters']]
```

```python
df_model.rename(columns={'Rating':'company_rating','Size':'company_size','Type of ownership':'type_of_ownership','Industry':'industry','Revenue':'revenue','Competitors':'competitors'},inplace=True)
```

# Handling Ordinal Categorical Features

```python
size_map={'Unknown':0,'1 to 50 employees':1, '51 to 200 employees':2, '201 to 500 employees':3,'501 to 1000 employees':4,'1001 to 5000 employees':5,'5001 to 10000 employees':6,'10001 to 20000 employees':7 }
df_model['company_size_rank']=df_model['company_size'].map(size_map)
df_model.drop('company_size',axis=True,inplace=True)
```

```python
revenue_map={'Unknown/Non-Applicable':0,'Less than $1 million (USD)':1,'$1 to $5 million (USD)':2,'$5 to $10 million (USD)':3,'$10 to $25 million (USD)':4,'$25 to $50 million (USD)':5, '$50 to $100 million (USD)':6,'$100 to $500 million (USD)':7,'$500 to $1 Billion (USD)':8,'$1 to $2 billion (USD)':9,'$2 to $5 billion (USD)':10,'$5 to $10 billion (USD)':11}
df_model['company_revenue_rank']=df_model['revenue'].map(revenue_map)
df_model.drop('revenue',axis=True,inplace=True)
```

```python
job_seniority_map={'na':0,'jr':1,'senior':2}
df_model['job_seniority_rank']=df_model['job seniority'].map(job_seniority_map)
df_model.drop('job seniority',axis=True,inplace=True)
```

```python
df_model.isnull().sum()
df_model.dropna(inplace=True)
```

```python
X=df_model.drop('average-salary',axis=1)
Y=df_model['average-salary']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
columns_to_scale=['company_rating','competitors','company_age','company_size_rank','company_revenue_rank']
X[columns_to_scale]=scaler.fit_transform(X[columns_to_scale])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
print("Training set size :{} and Testing set size: {}".format(X_train.shape,X_test.shape))
```

# Model Building

Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
```

```python
lr_model.fit(X_train,Y_train)
train_predictions = lr_model.predict(X_train)
test_predictions = lr_model.predict(X_test)

train_mse = mean_squared_error(Y_train, train_predictions)
test_mse = mean_squared_error(Y_test, test_predictions)

# Calculate the R-squared values
train_r_squared = lr_model.score(X_train, Y_train)
test_r_squared = lr_model.score(X_test, Y_test)

# Calculate "Accuracy-like" metric by converting R-squared to a percentage
train_accuracy_like = (1 - train_mse / Y_train.var()) * 100
test_accuracy_like = (1 - test_mse / Y_test.var()) * 100

print(f"Accuracy of the Linear Regression Model on Training set is: {train_accuracy_like:.2f}%")
print(f"Accuracy of the Linear Regression Model on Test set is: {test_accuracy_like:.2f}%")
```
